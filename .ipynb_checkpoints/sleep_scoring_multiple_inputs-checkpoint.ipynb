```python
EEG_files = ['../../code/LabCode/Fiber_photometry/mouse_653/mouse_653_230316.edf']
            # '../../code/LabCode/Fiber_photometry/mouse_807/mouse_807_230511.edf', 
            # '../../code/LabCode/Fiber_photometry/mouse_838/mouse_838_230525.edf', 
            # '../../code/LabCode/Fiber_photometry/mouse_994/mouse_994_230713.edf', 
            # '../../code/LabCode/Fiber_photometry/mouse_998/mouse_998_230727.edf', 
            # '../../code/LabCode/Fiber_photometry/mouse_1002/mouse_1002_230725.edf']
scores_files = ['../../code/LabCode/Fiber_photometry/mouse_653/mouse_653_230316_scores.tsv']
                # '../../code/LabCode/Fiber_photometry/mouse_807/mouse_807_230511_scores.tsv',
                # '../../code/LabCode/Fiber_photometry/mouse_838/mouse_838_230525_scores.tsv', 
                # '../../code/LabCode/Fiber_photometry/mouse_998/mouse_998_230727_scores.tsv', 
                # '../../code/LabCode/Fiber_photometry/mouse_994/mouse_994_230713_scores.tsv', 
                # '../../code/LabCode/Fiber_photometry/mouse_1002/mouse_1002_230725_scores.tsv']
epoch_limits = [8640]
                # 8640,
                # 8640,
                # 8640,
                # 8640,
                # 8640]
```

```python
CH1 = []
CH2 = []
CH3 = []
scores_converted = []
for EEG,scores_l,limits in zip(EEG_files,scores_files,epoch_limits):
    raw = mne.io.read_raw_edf(EEG, preload=True, verbose = False)
    sf = raw.info['sfreq']
    print('Data shape (channels, times) =', raw._data.shape)
    scores = pd.read_csv(scores_l, skiprows=10, delim_whitespace=True)
    scores_array = scores['Time.2']
    for i in range(limits): 
        CH1.append(raw._data[0][i*1280:(i+1)*1280])
        CH2.append(raw._data[1][i*1280:(i+1)*1280])
        CH3.append(raw._data[2][i*1280:(i+1)*1280])
    # in sirenia sleep, 255 unscored, 1 wake, 2 NREM, 3 REM, 4 "parameters"
    # in Yasa -2 = Unscored, -1 = Artefact / Movement, 0 = Wake, 1 = N1 sleep, 2 = N2 sleep, 3 = N3 sleep, 4 = REM sleep
    # for sleep scoring code -2 = Unscored, -1 = Artefact / Movement, 0 = Wake, 1 = N1 sleep, 1 = N2 sleep, 1 = N3 sleep, 2 = REM sleep
        if scores_array[i]==1: # wake
            scores_converted.append(0)
        elif scores_array[i]==2: # NREM
            scores_converted.append(1)
        elif scores_array[i]==3: # REM
            scores_converted.append(2)
        elif scores_array[i]==255: # unscored
            scores_converted.append(-2)      
array_scores = np.array(scores_converted).reshape(sum(epoch_limits),1)
scores_list = array_scores.tolist()
```

```python
total_epochs = sum(epoch_limits) 
sixty = round(total_epochs*0.6)
forty = round(total_epochs*0.4)

# train_CH1 = CH1[0:sixty]
# train_CH2 = CH2[0:sixty]
# train_CH3 = CH3[0:sixty]
# train_scores = scores_list[0:sixty]

# test_CH1 = CH1[sixty:]
# test_CH2 = CH2[sixty:]
# test_CH3 = CH3[sixty:]
# test_scores = scores_list[sixty:]

list_of_numbers = list(range(total_epochs))

train_CH1 = []
train_CH2 = []
train_CH3 = []
train_scores = []
for i in range(sixty): 
    num = list_of_numbers.pop(randrange(len(list_of_numbers)))
    train_CH1.append(CH1[num])
    train_CH2.append(CH2[num])
    train_CH3.append(CH3[num])
    train_scores.append(scores_list[num])
     
test_CH1 = []
test_CH2 = []
test_CH3 = []
test_scores = []
for i in range(forty): 
    num = list_of_numbers.pop(randrange(len(list_of_numbers)))
    test_CH1.append(CH1[num])
    test_CH2.append(CH2[num])
    test_CH3.append(CH3[num])
    test_scores.append(scores_list[num])
```

```python
scaler = RobustScaler()
scaled_train_CH1 = scaler.fit_transform(train_CH1)
scaled_train_CH2 = scaler.fit_transform(train_CH2)
scaled_train_CH3 = scaler.fit_transform(train_CH3)
scaled_test_CH1 = scaler.fit_transform(test_CH1)
scaled_test_CH2 = scaler.fit_transform(test_CH2)
scaled_test_CH3 = scaler.fit_transform(test_CH3)
```

```python
epoch = 4
fig, axs = plt.subplots(nrows=3,ncols=1,figsize=(10,5))
axs[0].plot(scaled_test_CH1[epoch])
axs[1].plot(scaled_test_CH2[epoch])
axs[2].plot(scaled_test_CH3[epoch])
for x in range(3):
    axs[x].set_ylim(-5,5)
plt.show()
print(test_scores[epoch])
```

```python
x_train = []
for i in range(len(scaled_train_CH1)):
    temp = []
    for j in range(1280):
        temp.append([scaled_train_CH1[i][j],scaled_train_CH2[i][j],scaled_train_CH3[i][j]])
    x_train.append(temp)
```

```python
X_train = np.array(x_train)
print(X_train.shape)
```

```python
x_test = []
for i in range(len(scaled_test_CH1)):
    temp = []
    for j in range(1280):
        temp.append([scaled_test_CH1[i][j],scaled_test_CH2[i][j],scaled_test_CH3[i][j]])
    x_test.append(temp)
```

```python
X_test = np.array(x_test)
print(X_test.shape)
```

```python
y_temp_train = []
for score in train_scores:
    if score == [0]:
        y_temp_train.append([1,0,0])
    if score == [1]:
        y_temp_train.append([0,1,0])
    if score == [2]:
        y_temp_train.append([0,0,1])
```

```python
model = Sequential()
model.add(
    layers.Bidirectional(
      layers.LSTM(
          units=15, 
          input_shape=[X_train.shape[1], X_train.shape[2]]
      )
    )
)
model.add(layers.Dropout(rate=0.5))
model.add(layers.Dense(units=15, activation='relu'))
model.add(layers.Dense(y_train.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.1,
    shuffle=True
)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend();
```

```python
model.evaluate(X_test, y_test)
```

```python
def get_uncompiled_model():
    input_EEG1 = Input(shape=(1280,), name = 'EEG1')
    input_EEG2 = Input(shape=(1280,), name = 'EEG2')
    input_EMG = Input(shape=(1280,), name = 'EMG')
    inputs = layers.concatenate([input_EEG1, input_EEG2, input_EMG], name = 'combined_inputs')
    x = layers.Dense(15, activation = "relu", name = 'layer_1')(inputs)
    x = layers.Dense(10, activation = "relu", name = 'layer_2')(x)
    outputs = layers.Dense(3, activation = "linear", name = 'outputs')(x)
    model = Model(inputs = [input_EEG1, input_EEG2, input_EMG], outputs = [outputs], name = 'sleep_scoring_model')
    return model

def get_compiled_model():
    model = get_uncompiled_model()
    model.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001),#0.001
    metrics = ["accuracy"],
    )
    return model
```

```python
model = get_compiled_model()
```
